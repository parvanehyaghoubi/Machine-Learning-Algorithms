```python
df = pd.read_csv('./datasets/cleaned_data_Task1.csv')
y = np.array(df.iloc[54][1: ],dtype = np.float64)
x = np.arange(1960, 2023)
```

```python
def cublog(x, a, b, c, d):
    return np.log(a*x**3 + b*x**2 + c*x + d)
```

```python
xdata= (x-x.min()) / (x.max() - x.min())
ydata= (y-y.min()) / (y.max() - y.min())

from scipy.optimize import curve_fit
popt,_ = curve_fit(cublog, xdata, ydata)
print(f" a = {popt[0]}, b = {popt[1]}")
```

```python
x_reg = np.arange(0.02,  1.03, 0.01)
y_reg = cublog(x_reg, *popt)
```

```python
x2025 = (2025 - x.min()) / (x.max() - x.min())

plt.scatter(xdata, ydata, color='crimson', label='Populaton of Czechia')
plt.plot(x_reg, y_reg, lw=3, label='Predicted', color='black')
plt.scatter(x2025, cublog(x2025 , *popt), color='blue', label='2025')
plt.xlabel('Year')
plt.ylabel('Czechia')
plt.legend(loc = 'upper left')
plt.show()
```
